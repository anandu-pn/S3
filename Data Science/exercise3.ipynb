```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#qn 1
x1=np.array([1,2,6,18])
y1=np.array([3,10,12,20])
plt.plot(x1,y1,'ro:',markerfacecolor='green',markeredgecolor='green',markersize=10)
plt.show()
```

```python
#2
temp=np.array([12,14,16,18,20,22,24])
sales=np.array([100,200,250,400,300,450,500])
plt.plot(temp,sales,'bo:')
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.show()
```

```python
with open('ex3.txt','r+') as file1:
    content=file1.read()
    xfile,yfile=content.split("\n")
    xfile=list(map(int,xfile.split(",")))
    yfile=list(map(int,yfile.split(",")))
    plt.plot(xfile,yfile,"yo:")
    plt.xlabel("Temp")
    plt.ylabel("Sales")
    plt.title("Temp vs Sales")
    plt.show()
```

```python
#plot two more lines
line1x=np.array([1,2,3,4,5])
line1y=np.array([2,5,8,12,15])
line2y=np.array([2,4,6,8,10])
plt.plot(line1x,line1y,'bo:',label="Data 1")
plt.plot(line1x,line2y,'yo:',label="Data 2")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'g')  # green line
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'b:')  # blue line
plt.title("Cosine Wave")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.show()
```

```python
#6 popularoty of languages
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7, 6.7]

plt.bar(languages, popularity,color="green")
plt.title("Popularity of Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Popularity (%)")
plt.show()
#horizontal sub qn 2
plt.barh(languages, popularity, color='red')
plt.title("Popularity of Programming Languages (Horizontal)")
plt.xlabel("Popularity (%)")
plt.ylabel("Languages")
plt.show()
#multi color qn
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
plt.bar(languages, popularity, color=colors)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages - Colored Bars')
plt.show()
```

```python
men_means = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)
groups = ('G1', 'G2', 'G3', 'G4', 'G5')
x = np.arange(len(groups)) 
bar_width = 0.35 
plt.bar(x - bar_width/2, men_means, bar_width, label='Men', color='green')
plt.bar(x + bar_width/2, women_means, bar_width, label='Women', color='red')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.xticks(x, groups)
plt.legend()

plt.show()
```

```python
with open("ex3.csv","r+") as medals:
    lines=medals.readlines()[1:]
    xpie=[]
    ypie=[]
    for line in lines:
        x=line.strip().split(",")
        xpie.append(x[0])
        ypie.append(int(x[1]))
    plt.pie(ypie,labels=xpie,startangle=90)
    plt.title("Top 5 Gold Medal")
    plt.show()
```
